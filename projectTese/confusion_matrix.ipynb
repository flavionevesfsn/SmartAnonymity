# Matriz de Confus√£o da base de dados com 121 DataSets

```python
from sklearn.metrics import confusion_matrix, classification_report
from pandas import read_csv

import numpy as np
from scipy import stats

from sklearn import datasets
from sklearn.semi_supervised import LabelSpreading

df = read_csv('datasets/outputSummaryOutputData-matriz.csv')
y_true, y_pred = df['y_true'], df['y_pred']

report_matrix = confusion_matrix(y_true, y_pred, labels=["Agriculture", "HealthCare", "IndependentLiving", "IndustrialProcesses",
                                         "Logistic", "SecurityPublic", "SmartCampus", "SmartGrid", "SmartHome",
                                         "Undefined", "UrbanMobility"])

print(report_matrix)
```

```python
#Visualiza√ß√£o Gr√°fica da Matriz de Confus√£o
import matplotlib.pyplot as plt
import seaborn as sns

#matriz_title = 'Decision Tree - Validation'
fig = plt.figure(figsize=(10,8))
fig.suptitle('Confusion Matrix', fontsize=14, fontweight='bold')

sns.heatmap(report_matrix, cmap='Blues', annot=True)

#plt.title(matriz_title, fontsize=14)

plt.xticks([])
plt.yticks([])
```
